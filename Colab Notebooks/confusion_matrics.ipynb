```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
plt.rcParams['figure.figsize'] = (16, 7)
#
```

```python
df = pd.read_csv('titanic_cleaned.csv')
print(df.head())
print(df.shape)
```

```python
x = df.drop('Survived', axis=1)
y = df['Survived']
#
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=101)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression().fit(x_train, y_train)
#
```

```python
lr_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

print('++++++++++++++++++++++++++++++++')

lr_pred_accuracy = round(accuracy_score(y_test, lr_pred)*100, 2)

print(f'Logistic Regression Accuracy: {lr_pred_accuracy}%')

print('++++++++++++++++++++++++++++++++')
```

# Confusion Metrics

```python
from sklearn.metrics import confusion_matrix, recall_score,precision_score

precision = precision_score(y_test, lr_pred)

recall = recall_score(y_test, lr_pred)

confusion_matrix = confusion_matrix(y_test,lr_pred)

print(f'Precision: {precision:.2f}')

print(f'Recall: {recall:.2f}')

print(f'Confusion Matrix:{confusion_matrix}')

sns.heatmap(confusion_matrix,annot = True,cmap='Blues')
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier().fit(x_train,y_train)

dt_pred = dt.predict(x_test)

dt_pred_accuracy = round(accuracy_score(y_test,dt_pred) * 100, 2)

print(f'Decision Tree Accuracy: {dt_pred_accuracy}%')

from sklearn.metrics import confusion_matrix, recall_score,precision_score

precision = precision_score(y_test,dt_pred)

recall = recall_score(y_test,dt_pred)

confusion_matrix = confusion_matrix(y_test,dt_pred)

print(f'Precision: {precision:.2f}')

print(f'Recall: {recall:.2f}')

print(f'Confusion Matrix:{confusion_matrix}')

sns.heatmap(confusion_matrix,annot = True)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier().fit(x_train,y_train)

rfc_pred = rfc.predict(x_test)

rfc_pred_accuracy = round(accuracy_score(y_test,rf_pred ) * 100,2)

print(f'Random Forest Classifier Accuracy: {rfc_pred_accuracy}%')

# confusion matrix

from sklearn.metrics import confusion_matrix , recall_score , precision_score

precision = precision_score(y_test , rfc_pred)

recall = recall_score(y_test , rfc_pred)

confusion = confusion_matrix(y_test , rfc_pred)

print(f'Precision : {precision:.2f}')
print(f'Recall : {recall:.2f}')
print(f'Confusion Matrix : {confusion}')

sns.heatmap(confusion , annot = True)
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
```

# Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier

gbc = GradientBoostingClassifier().fit(x_train,y_train)

gbc_pred = gbc.predict(x_test)

gbc_pred_accuracy = round(accuracy_score(y_test,gbc_pred) * 100, 2)

print(f'Gradient Boosting Classifier Accuracy: {gbc_pred_accuracy}%')

from sklearn.metrics import confusion_matrix, recall_score, precision_score

precision = precision_score(y_test,gbc_pred)

recall = recall_score(y_test,gbc_pred)

confusion_matrix = confusion_matrix(y_test,gbc_pred)

print(f'Precision: {precision:.2f}')

print(f'Recall: {recall:.2f}')

print(f'Confusion Matrix:{confusion_matrix}')

sns.heatmap(confusion_matrix,annot = True)
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
```
