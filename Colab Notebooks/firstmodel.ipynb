```python
# prompt: what is quartile range and how to handle outliers

# Calculate the Interquartile Range (IQR) for the 'Age' column
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

print(Q1,Q3,IQR)
# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
lower_bound, upper_bound
lower_bound = max(0,lower_bound)
print(lower_bound)
df.Age.min()
```

```python
df.describe()
```

```python

# Identify outliers
outliers = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]
print(outliers)
print(df.Age.max())
```

```python
# Handle outliers
# Option 1: Remove outliers
df_no_outliers = df[(df['Age'] >= lower_bound) & (df['Age'] <= upper_bound)]
print(df_no_outliers.Age.min())
```

```python

# Option 2: Replace outliers with the upper or lower bound
df['Age'] = np.where(df['Age'] > upper_bound, upper_bound, df['Age'])
df['Age'] = np.where(df['Age'] < lower_bound, lower_bound, df['Age'])
df.Age.max(),df.Age.min()
```

```python
def impute_age(cole):
    Age = cole[0]
    Pclass = cole[1]
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age

#apply funcation
df['Age'] = df[['Age','Pclass']].apply(impute_age , axis = 1)
print(df.Age.min())
df.Age.max()
```
