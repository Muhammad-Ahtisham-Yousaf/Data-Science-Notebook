```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data_Lec13.csv",encoding = 'latin1')
```

```python
df.head(5)
```

```python
df.shape
```

```python
df = df.rename(columns = {"Engine HP":"HP","Transmission Type":"Transmissioin","Engine Cylinders":"Cylinders","MSRP":"Price"})
df.tail()
```

```python
# df = df.drop(["Vehicle Size"],axis = 1)
# df = df.drop(["Make"],axis = 1)
# df = df.drop(["Model"],axis = 1)
df = df.drop(["Engine Fuel Type"],axis = 1)
df = df.drop(["Driven_Wheels"],axis = 1)

df = df.drop(["Market Category"],axis = 1)
df = df.drop(["Vehicle Style"],axis = 1)

df.head()
```

```python
df = df.drop(["Transmissioin"],axis = 1)
```
