```python
import pandas as pd
import wbdata
```

```python
indicator = {
    'NY.GDP.MKTP.KD.ZG':'GDP growth (annual %)'
}
countries = ['IND','PAK',"USA"]
date = ('2015','2024')
```

```python
wbdata.get_countries()
```

```python
wbdata.get_sources()
```

```python
wbdata.get_indicators(source = 25)
```

```python
wbdata.get_data('NY.GDP.PCAP.PP.KD',country = countries)
```

```python
# NY.GDP.MKTP.KD.ZG     GDP growth (annual %)
# NY.GDP.PCAP.KD        GDP per capita (constant 2015 US$)
# NY.GDP.PCAP.PP.KD     GDP per capita, PPP (constant 2021 international $)

wbdata.get_data('NY.GDP.PCAP.KD',country =[ 'PAK','IND',"USA"],date = ("2015","2024"))
```

```python
data = wbdata.get_dataframe(indicator,country = countries,date = date)
```

```python
print(data)
```

```python
data.reset_index()
```

```python
data.to_csv('GDP2.csv')
```
