```python
# for single feature
def replace_outlier_with_median(train, column):
    # Calculate Q1, Q3, and IQR
    Q1 = train[column].quantile(0.25)
    Q3 = train[column].quantile(0.75)
    IQR = Q3 - Q1

    # Define the lower and upper boundaries
    lower_boundary = Q1 - 1.5 * IQR
    upper_boundary = Q3 + 1.5 * IQR

    # Calculate the median of the column
    median_value = train[column].median()

    # Replace outliers with the median
    train[column] = train[column].apply(lambda x: median_value if x < lower_boundary or x > upper_boundary else x)

    return train
```

```python
# function to process outliers and nulls
def process_outliers_and_nulls(test):
    for column in test.columns:
        # if column is numerical
        if pd.api.types.is_numeric_dtype(test[column]):
            # Calculate IQR
            Q1 = test[column].quantile(0.25)
            Q3 = test[column].quantile(0.75)
            IQR = Q3 - Q1

            #Define outliers boundries
            lower_boundry = Q1 - 1.5 * IQR
            upper_boundry = Q3 + 1.5 * IQR

            #replace outliers with median
            median_value = test[column].median()
            test[column] = np.where((test[column] < lower_boundry) | (test[column] > upper_boundry),median_value,test[column])
        if pd.api.types.is_object_dtype(test[column]):
            test[column].fillna(method = 'ffill',inplace = True)
    return test
```

```python
def fill_nulls(test):
    """
    Fill null values in numeric columns with median
    and in categorical columns with forward fill.

    Args:
    dataframe (pd.DataFrame): Input DataFrame with null values.

    Returns:
    pd.DataFrame: DataFrame with null values filled.
    """
    for col in test.columns:
        if pd.api.types.is_numeric_dtype(test[col]):
            # Replace nulls in numeric columns with median
            median_value = test[col].median()
            test[col].fillna(median_value, inplace=True)
        elif pd.api.types.is_string_dtype(test[col]):
            # Replace nulls in categorical columns with forward fill
            test[col].fillna(method='ffill', inplace=True)


```
