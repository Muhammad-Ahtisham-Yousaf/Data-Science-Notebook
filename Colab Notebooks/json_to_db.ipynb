```python
# prompt: concat my all data frames in a single dataframe with thier data using id

import pandas as pd

# Assuming checkin_df, business_df, review_df, tip_df, and user_df are already defined as in your previous code

# Create a list to hold all dataframes
dfs = [checkin_df, business_df, review_df, tip_df, user_df]

# Concatenate the dataframes
try:
  # Attempt to concatenate using the 'id' column. If 'id' column exists in all dataframes
  all_data = pd.concat(dfs, ignore_index=True)
except KeyError:
    print("'id' column not found in all dataframes. Check for common identifier column and modify code")

    # Handle the case where 'id' column does not exist in all DataFrames
    # Find a common column across the DataFrames and use that as the key
    # For example:
    common_column = None
    for df in dfs:
        for col in df.columns:
            if col in dfs[0].columns and col in df.columns:
                common_column = col
                break

    if common_column:
        all_data = pd.concat(dfs, ignore_index=True)
    else:
        print("No common column across dataframes. Concatenation will be done with ignore_index=True which will reset the index.")
        all_data = pd.concat(dfs, ignore_index=True)  # Concatenate without a key


# Check if the id column is duplicated and count the duplicates
id_duplicates = all_data.duplicated(subset=['business_id'], keep=False)
print(f"Number of duplicates: {id_duplicates.sum()}")

# Print the concatenated dataframe's head
print(all_data.head())
```

```python
import pandas as pd
import json
import numpy as np
from sqlalchemy import create_engine
```

```python
with open('/content/yelp-dataset/yelp_academic_dataset_business.json', 'r') as f:
    business_data = [json.loads(line) for line in f]
data_df = pd.DataFrame(business_data)

with open('/content/yelp-dataset/yelp_academic_dataset_checkin.json', 'r') as f:
    checkin_data = [json.loads(line) for line in f]
checkin_df = pd.DataFrame(checkin_data)

with open('/content/yelp-dataset/yelp_academic_dataset_review.json', 'r') as f:
    review_data = [json.loads(line) for line in f]
review_df = pd.DataFrame(review_data)

with open('/content/yelp-dataset/yelp_academic_dataset_tip.json', 'r') as f:
    tip_data = [json.loads(line) for line in f]
tip_df = pd.DataFrame(tip_data)

with open('/content/yelp-dataset/yelp_academic_dataset_user.json', 'r') as f:
    user_data = [json.loads(line) for line in f]
user_df = pd.DataFrame(user_data)
```

```python
data_df.drop(['attributes'],axis = 1,inplace = True)
```

```python
data_df.drop(['hours'],axis = 1,inplace = True)
```

```python
engine = create_engine('sqlite:///yelp.db')

def load_df_to_db(df, table_name,engine):
    df.to_sql(table_name,con = engine, if_exists='replace', index=False)

load_df_to_db(data_df,'business',engine)
load_df_to_db(checkin_df,'checkin',engine)
load_df_to_db(review_df,'review',engine)
load_df_to_db(tip_df,'tip',engine)
load_df_to_db(user_df,'user',engine)
```
