```python
def generate_blog(topic):
    messages = [
        {"role":"system","content":"""You are trained to analyze a topic a topic and generate a blog post.
                                   The blog post must contain 1000 to 1200 words (No less than 700 words)."""},
        {"role":"user","content":f"""Analyze the topic and generate a blog post.The topic is {topic}
                                 The blog post should contain the following format.
                                1) Title (not more than one line)
                                2) Introduction (Give intro about topic).
                                3) Add an image url relevent to the topic.
                                4) Add 3/2 subheadings and explain them.
                                5) Body (should describe the facts and findings).
                                6) Conclusion of the topic.

                                7) general FAQ about the topic.
                                """}
    ]
    response = openai.ChatCompletion.create(
        model = "gpt-3.5-turbo",
        messages = messages,
        max_tokens = 1500,
        n= 1,
        stop = None,
        temperature = 0.5
    )
    response_text = response.choices[0].messages['content'].strip().lower()
    return response_text

topic = "Artificial Intelligence"
blog = generate_blog(topic)
print(blog)
```

```python
# prompt: generate a paragraph of 50 words

def generate_paragraph(num_words):
    messages = [
        {"role": "system", "content": "You are a helpful assistant that generates creative text."},
        {"role": "user", "content": f"Generate a paragraph with approximately {num_words} words."}
    ]
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=messages,
        max_tokens=150,
        n=1,
        stop=None,
        temperature=0.5
    )
    response_text = response.choices[0].message['content'].strip()
    return response_text

paragraph = generate_paragraph(50)
print(paragraph)
```
