```python
# preprocessing
from sklearn import preprocessing

# non_numeric_var = x.select_dtypes(exclude = ['number']).columns

# x_numeric = x.drop(non_numeric_var,axis = 1)

# x_numaric = x_numeric.values

pre_process = preprocessing.StandardScaler().fit(x)

x_transform = pre_process.fit_transform(x)
```

```python
x_transform
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_transform,y, test_size = .10 ,random_state=101)
```

# Linear Regression

```python
# import Model
from sklearn.linear_model import LinearRegression

# creat instance of model
my_regressor = LinearRegression()

#pass training data into model
my_regressor.fit(x_train,y_train)
```

# prediction

```python
print(x_test.shape)
x_test
# print(x_train.shape)
# print(y_train.shape)
# y_train
# print(y_test.shape)
```

```python
y_pred = my_regressor.predict(x_test)
print(y_pred.shape)
```

```python
sns.scatterplot(x = y_test,y = y_pred, color = "pink",label = "Actual data points")
plt.plot([min(y_test),max(y_test)],[min(y_test),max(y_test)],color = 'red',label = 'Regression line')
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
```

```python
results = np.column_stack((y_test,y_pred))

print("Actual Values  ")
for actual,pridected in results:
    print(f'{actual:14.2f},{pridected:12.2f}')
```
