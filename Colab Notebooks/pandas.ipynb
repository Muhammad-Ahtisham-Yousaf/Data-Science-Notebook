```python
import pandas as pd
df = pd.read_csv("Salaries.csv")
df.head(10)
```

```python
df.drop(['Notes','Status','TotalPayBenefits'],axis = 1,inplace = True)
```

```python
df.head(10)
```

```python
df.Benefits.first_valid_index()
```

```python
df.describe()
```

```python
columns_to_fix = ['OvertimePay', 'OtherPay', 'TotalPay', 'BasePay', 'Benefits']

for col in columns_to_fix:
    # For 'OvertimePay', set negative values to 0
    if col == 'OvertimePay':
        neg_otp = df[df[col] < 0]
        neg_count = neg_otp[col].count()
        neg_indeces = neg_otp.index
        df.loc[neg_indeces, col] = 0  # Replacing negative values with 0

        print("Negative Counts of 'OvertimePay':", neg_count)
        print('Negative Indices of "OvertimePay Feature":', neg_indeces.tolist())

    # For 'TotalPay' and 'BasePay', replace values <= 1 with the column mean
    elif col == 'TotalPay':
        neg_tp = df[df[col] <= 1]
        neg_count = neg_tp[col].count()
        neg_indeces = neg_tp.index
        df.loc[neg_indeces, col] = df[col].mean()

        print("Negative Counts of 'TotalPay':", neg_count)
        print('Negative Indices of "TotalPay Feature":', neg_indeces.tolist())

    elif col == 'BasePay':
        neg_bp = df[df[col] <= 1]
        neg_count = neg_bp[col].count()
        neg_indeces = neg_bp.index
        df.loc[neg_indeces, col] = df[col].mean()

        print("Negative Counts of 'BasePay':", neg_count)
        print('Negative Indices of "BasePay Feature":', neg_indeces.tolist())

    # For 'OtherPay' and 'Benefits', set negative values to 0
    elif col == 'OtherPay':
        neg_op = df[df[col] < 0]
        neg_count = neg_op[col].count()
        neg_indeces = neg_op.index
        df.loc[neg_indeces, col] = 0  # Replacing negative values with 0

        print("Negative Counts of 'OtherPay':", neg_count)
        print('Negative Indices of "OtherPay Feature":', neg_indeces.tolist())

    elif col == 'Benefits':
        neg_bnft = df[df[col] < 0]
        neg_count = neg_bnft[col].count()
        neg_indeces = neg_bnft.index
        df.loc[neg_indeces, col] = 0  # Replacing negative values with 0

        print("Negative Counts of 'Benefits':", neg_count)
        print('Negative Indices of "Benefits Feature":', neg_indeces.tolist())


```
