```python
# prompt: how to add text on plot like smoker non smoker using seaborn

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
df = sns.load_dataset("tips")

# Create the count plot
ax = sns.countplot(x='sex', data=df, hue='smoker', saturation=0.7)

# Add text annotations to the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

# Add a legend and labels
plt.ylabel("People")
plt.legend(title="Smoker Status")
plt.show()
```

# Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action = 'ignore')
# Load the tips dataset
df = sns.load_dataset('tips')

# Create a pivot table for the heatmap (average tip by day and time)
pivot_table = df.pivot_table(values='total_bill', index='day', columns='time', aggfunc='max')

print(pivot_table)

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title("Average Tip by Day and Time")
plt.show()
```
